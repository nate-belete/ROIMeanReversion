# Example of how to use the ROIMeanReversion class

This example will demonstrate how to use the ROIMeanReversion class to calculate the return on investment for a given stock.


## Import packages

First, we need to import the necessary packages.


```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Import the ROIMeanReversion class

Next, we need to import the ROIMeanReversion class.


```python
from ROIMeanReversion import ROIMeanReversion
```

## Create an instance of the ROIMeanReversion class

Now, let's create an instance of the ROIMeanReversion class. We will use the ticker 'AAPL' and the variable 'Close' for our example. We will also use a rolling window size of 20 and retrieve data from 2018-01-01 to 2023-12-31 in intervals of 1 week.



```python
# Create instance of the ROIMeanReversion class
roi_mean_reversion = ROIMeanReversion(ticker='KHC', 
                                      var='Close', 
                                      rolling_window_size=7, 
                                      fast_sma = 9,
                                      slow_sma = 20,
                                      start_date='2010-01-01',
                                      end_date='2023-12-31', 
                                      interval='1wk'
                                     )
```

## Calculate the return on investment

Now, let's calculate the return on investment for the given stock.


```python
# Calculate ROI
roi_mean_reversion.calculate_ROI()
```

## Calculate the rolling window mean of the ROI

Next, we need to calculate a rolling window mean of the ROI.


```python
# Calculate rolling window mean of ROI
roi_mean_reversion.calculate_N_period_ROI_MA()
```

## Calculate the Fast and Slow Simple Moving Averages

Next, we need to calculate the fast and slow SMA's.


```python
roi_mean_reversion.calculate_SMA_MA()
```

## Create buy and sell conditions

Now, let's create buy and sell conditions based on the ROI mean.


```python
# Create buy and sell conditions
roi_mean_reversion.create_buy_sell_conditions()
```

## Plot the results

Next, let's plot the stock data and buy/sell signals.

```python
# Plot results
roi_mean_reversion.plot_results()
```

## Buy and sell stock

Finally, let's buy and sell the stock based on the action.

```python
# Buy and sell stock
roi_mean_reversion.buy_sell_stock()
```

And that's it! We have successfully used the ROIMeanReversion class to analyze a stock and create a buy/sell strategy.
